```python
#let´s create a new dictionary where each key is a unique app name and the corresponding 
# dictionary value is the highest number of reviews of that app
reviews_max = { }
for app in google_data:
    name = app[0]
    n_reviews = float(app[3])
    
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
        
    elif name not in reviews_max:
        reviews_max[name] = n_reviews
print(len (reviews_max))
```

* We start by initializing two empty lists, android_clean and already_added.
* We loop through the android data set, and for every iteration:
  * We isolate the name of the app and the number of reviews.
  * We add the current row (app) to the android_clean list, and the   app name (name) to the already_added list if:
    * The number of reviews of the current app matches the number of reviews of that app as described in the reviews_max dictionary; and
    * The name of the app is not already in the already_added list. We need to add this supplementary condition to account for those cases where the highest number of reviews of a duplicate app is the same for more than one entry (for example, the Box app has three entries, and the number of reviews is the same). If we just check for reviews_max[name] == n_reviews, we'll still end up with duplicate entries for some apps.

```python
#now we will remove  duplicate rows
android_clean=[] # store new cleaned data set
already_added=[] # store app names

for app in google_data:
    name = app[0]
    n_reviews = float(app[3])
    
    if (reviews_max[name] == n_reviews) and (name not in already_added):
        android_clean.append(app)
        already_added.append(name)
```

## REMOVING NON-ENGLISH APPS

As our interest is left apps which are directed toward an English-speaking audience. 
one way to remove those apps that we are not interesting in is those that their name contain a symbol which is not coomonly used in Engligh text.
each character we use in string has a number associated with it. For instance, the number for the character "a" is 97 and chaaracter "A" is 65, and for the character is '爱' is 29233.

```python
print(ord("a"))
print(ord("A"))
print(ord('爱'))
```

the numbers corresponding to the characters commonly used in an english text are in the range from 0 to 127, according to [ASCII](https://en.wikipedia.org/wiki/ASCII) System. based on the number range if tje number is equal to or lees tjan 127, then the character belongs to the set of common English characters. Otherwise, if any character is greater than 127 propably means that the apps has a non-english name.

```python
# we define a function which return False if input is > 127 
# or True if input is < 127
def convert(string):
    
    for character in string:
        if ord(character) > 127:
            return False
        else:
            return True
```

```python
#let´s test our function
print(convert('Instagram'))
print(convert('爱奇艺PPS -《欢乐颂2》电视剧热播'))
print(convert('Docs To Go™ Free Office Suite'))
print(convert('Instachat 😜'))
```

```python
# we create frequency table for data set in order to obtain  
# the number of each category and genre that is repeated
# also we combine this function with other in order to display columns
# in descending order for a better visualization
def freq_table(dataset,index):
    frequency_table={}
    
    for row in dataset:
        category= row[index]
        if category in frequency_table:
            frequency_table[category] += 1 / len(dataset)*100
        else:
            frequency_table[category] =1
    return frequency_table

def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```

## ANDROID DATA ANALYSIS

Once we have analysed iOS data now proceed to check android data with the number of installs.

```python
display_table(android,5) # shows installs column
```

Quantity of installs is set by ranges of installs number so we cannot get exact information, we don't know whether an app with 100,000+ installs has 100,000 installs, 200,000, or 350,000. However, we do not need a precise data we just need to know which app category attract the most users.
We will consider numbers as they are, for instance an ap with 100,000 + will be considered that app has 100,000 installs. for this it is neccesary to convert each install number from string to float, commas will be removed ans the plus characters in order to avoid errors.

```python
category_android=freq_table(android,1)

for category in category_android:
    total = 0
    len_category = 0
    for app in android:
        category_app = app[1]
        if category_app == category:            
            n_installs = app[5]
            n_installs = n_installs.replace(',', '')
            n_installs = n_installs.replace('+', '')
            total += float(n_installs)
            len_category += 1
            
    avg_n_installs = total / len_category
    print(category, ':', avg_n_installs)
```
